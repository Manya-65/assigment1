## Hyperparameter tuning

### Subtask:
Implement a hyperparameter tuning strategy (e.g., grid search or random search) to find optimal hyperparameters for the ANN model.

**Reasoning**:
Implement hyperparameter tuning using Keras Tuner's RandomSearch. Define the `build_model` function for the tuner, instantiate the tuner object, and start the search.

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Define the input shape and number of classes outside the build_model function
input_shape = (X_train.shape[1],)
num_classes = y_train_one_hot.shape[1]

def build_model(hp):
    """Builds a sequential model for Keras Tuner."""
    model = Sequential()
    model.add(Dense(units=hp.Int('units_input', min_value=32, max_value=128, step=32),
                    activation='relu',
                    input_shape=input_shape))
    model.add(Dense(units=hp.Int('units_hidden', min_value=32, max_value=128, step=32),
                    activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(optimizer=Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Instantiate the Keras Tuner RandomSearch object
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=2,
    directory='my_dir',
    project_name='alphabet_classification'
)

# Print a summary of the search space
print("Search space summary:")
tuner.search_space_summary()

# Start the hyperparameter search
print("\nStarting hyperparameter search...")
tuner.search(X_train, y_train_one_hot, epochs=10, validation_data=(X_test, y_test_one_hot))

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print("\nBest Hyperparameters found by the tuner:")
print(best_hps.values)
```
