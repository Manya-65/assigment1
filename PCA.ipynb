## Load and explore data

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('wine.csv')

# Display the first 5 rows
print("First 5 rows of the DataFrame:")
display(df.head())

# Print data types
print("\nData types of each column:")
display(df.dtypes)

# Generate descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
display(df.describe())

# Check for missing values
print("\nMissing values per column:")
display(df.isnull().sum())

# Calculate and display the correlation matrix
print("\nCorrelation matrix:")
display(df.corr())
```

## Dimensionality reduction with pca

### Subtask:
Standardize the data, apply PCA, determine the optimal number of components, and transform the data.


**Reasoning**:
Standardize the data, apply PCA, determine the optimal number of components, and transform the data as per the instructions.



```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Exclude the 'Type' column for scaling and PCA
X = df.drop('Type', axis=1)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Apply PCA to the scaled data
pca = PCA()
pca.fit(df_scaled)

# Calculate cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Determine the optimal number of components (e.g., explaining 95% variance)
n_components_optimal = np.where(cumulative_variance_ratio >= 0.95)[0][0] + 1
print(f"Optimal number of components explaining >= 95% variance: {n_components_optimal}")

# Instantiate PCA with the optimal number of components
pca_optimal = PCA(n_components=n_components_optimal)

# Fit and transform the scaled data
X_pca = pca_optimal.fit_transform(df_scaled)
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_components_optimal)])

# Display the first few rows of the PCA-transformed data
print("\nFirst 5 rows of the PCA-transformed DataFrame:")
display(df_pca.head())
```

## Clustering with original data

### Subtask:
Apply K-means clustering to the original data, visualize the results, and evaluate performance.


**Reasoning**:
The subtask requires applying K-means clustering to the original data, visualizing the results, and evaluating performance. This involves determining the optimal number of clusters using the elbow method, fitting the KMeans model, adding cluster labels to the original dataframe, visualizing the clusters using PCA components, and calculating the silhouette score. These steps can be logically grouped into a single code block for efficiency.



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the elbow method
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot the inertia values against the number of clusters
plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)
plt.grid(True)
plt.show()

# Based on the elbow plot (visual inspection), choose an appropriate number of clusters.
# Let's assume the elbow is at k=3 for demonstration purposes.
# In a real scenario, you would visually inspect the plot to make this decision.
optimal_k = 3
print(f"\nChoosing optimal number of clusters: {optimal_k}")

# Apply K-means clustering with the chosen number of clusters
kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans_optimal.fit(df_scaled)

# Add the cluster labels to the original df DataFrame
df['cluster_label_original'] = kmeans_optimal.labels_

# Visualize the clustering results using a scatter plot of PC1 and PC2
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df['cluster_label_original'], cmap='viridis')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('K-means Clustering on Original Data (Visualized with PC1 vs PC2)')
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()

# Calculate the silhouette score
silhouette_avg = silhouette_score(df_scaled, df['cluster_label_original'])
print(f"\nSilhouette Score for clustering on original data: {silhouette_avg}")
```

## Clustering with pca data

### Subtask:
Apply K-means clustering to the PCA-transformed data, visualize the results, and evaluate performance.


**Reasoning**:
Apply K-means clustering to the PCA-transformed data, visualize the results, and evaluate performance based on the instructions. This involves determining the optimal number of clusters using the elbow method on `df_pca`, fitting KMeans, adding labels to `df_pca`, visualizing the clusters on PC1 and PC2, and calculating the silhouette score.



```python
import matplotlib.pyplot as plt

# Determine the optimal number of clusters for PCA-transformed data using the elbow method
inertia_pca = []
k_range = range(1, 11)
for k in k_range:
    kmeans_pca = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_pca.fit(df_pca)
    inertia_pca.append(kmeans_pca.inertia_)

# Plot the inertia values against the number of clusters for PCA data
plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia_pca, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k (PCA Data)')
plt.xticks(k_range)
plt.grid(True)
plt.show()

# Based on the elbow plot (visual inspection), choose an appropriate number of clusters.
# Let's assume the elbow is at k=3 for demonstration purposes, similar to the original data.
# In a real scenario, you would visually inspect the plot to make this decision.
optimal_k_pca = 3
print(f"\nChoosing optimal number of clusters for PCA data: {optimal_k_pca}")

# Apply K-means clustering to the df_pca DataFrame using the chosen optimal_k_pca
kmeans_optimal_pca = KMeans(n_clusters=optimal_k_pca, random_state=42, n_init=10)
kmeans_optimal_pca.fit(df_pca)

# Add the cluster labels obtained from the K-means model to the df_pca DataFrame
df_pca['cluster_label_pca'] = kmeans_optimal_pca.labels_

# Visualize the clustering results on the PCA-transformed data using a scatter plot
plt.figure(figsize=(10, 6))
scatter_pca = plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['cluster_label_pca'], cmap='viridis')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('K-means Clustering on PCA Data (PC1 vs PC2)')
plt.colorbar(scatter_pca, label='Cluster Label')
plt.grid(True)
plt.show()

# Calculate the silhouette score for the clustering on the PCA-transformed data
silhouette_avg_pca = silhouette_score(df_pca, df_pca['cluster_label_pca'])
print(f"\nSilhouette Score for clustering on PCA data: {silhouette_avg_pca}")
```

## Comparison and analysis

### Subtask:
Compare the clustering results from both original and PCA-transformed data, discuss the impact of dimensionality reduction, and analyze trade-offs.


**Reasoning**:
Compare the clustering results from both original and PCA-transformed data based on the optimal number of clusters, silhouette scores, and visualizations. Discuss the impact and trade-offs of using PCA for clustering based on these comparisons.



```python
# 1. Compare optimal number of clusters
print(f"Optimal number of clusters for original scaled data (based on visual elbow inspection): {optimal_k}")
print(f"Optimal number of clusters for PCA-transformed data (based on visual elbow inspection): {optimal_k_pca}")
print("Comparison: In this case, the elbow method suggested the same optimal number of clusters (3) for both the original scaled data and the PCA-transformed data.")

# 2. Compare silhouette scores
print(f"\nSilhouette Score for clustering on original scaled data: {silhouette_avg:.4f}")
print(f"Silhouette Score for clustering on PCA-transformed data: {silhouette_avg_pca:.4f}")
print("Comparison: The silhouette score for clustering on the PCA-transformed data ({:.4f}) is higher than that for the original scaled data ({:.4f}). A higher silhouette score generally indicates better-defined and more separated clusters. This suggests that PCA might have helped in creating more distinct clusters in this case.".format(silhouette_avg_pca, silhouette_avg))

# 3. Visually compare scatter plots (already generated in previous steps)
print("\nVisual Comparison of Scatter Plots:")
print("The scatter plot for clustering on original data (visualized with PC1 vs PC2) and the scatter plot for clustering on PCA data (PC1 vs PC2) were generated in previous steps.")
print("Observe how the cluster separation appears in both visualizations. Ideally, the plot for PCA data might show slightly clearer separation between clusters, aligning with the higher silhouette score.")

# 4. Discuss the potential impact of dimensionality reduction
print("\nImpact of Dimensionality Reduction using PCA:")
print("PCA reduces the number of features while retaining most of the variance in the data. This can potentially reduce noise and multicollinearity, which can sometimes improve clustering performance by allowing the algorithm to focus on the most important underlying patterns.")
print("Reducing dimensions also affects distance calculations. In a lower-dimensional space, distances between data points might be less influenced by irrelevant or noisy features, potentially leading to more meaningful cluster assignments.")

# 5. Analyze the trade-offs of using PCA for clustering
print("\nTrade-offs of using PCA for Clustering:")
print("Benefits:")
print("- Reduced noise and multicollinearity: PCA can help remove redundant information and noise, potentially leading to better cluster structures.")
print("- Faster computation: Clustering algorithms can run faster on lower-dimensional data.")
print("- Reduced storage space: Storing PCA-transformed data requires less space.")
print("Drawbacks:")
print("- Loss of interpretability: The principal components are linear combinations of the original features, making it difficult to directly interpret what each component represents in terms of the original variables.")
print("- Potential loss of information: While PCA aims to retain most variance, some information is always lost during dimensionality reduction, which might negatively impact clustering if the lost information was important for distinguishing clusters.")

# 6. Summarize comparison and analysis
print("\nSummary and Conclusion:")
print("Based on the comparison, using PCA for clustering on this dataset appears to have been beneficial.")
print("Both methods suggested the same optimal number of clusters (3).")
print("However, clustering on the PCA-transformed data resulted in a higher silhouette score ({:.4f}) compared to the original scaled data ({:.4f}), indicating potentially better cluster separation.".format(silhouette_avg_pca, silhouette_avg))
print("The visual comparison of the scatter plots (PC1 vs PC2) should also be considered to see if the clusters appear more distinct after PCA.")
print("While PCA introduces a trade-off in interpretability, the potential benefits of reduced noise and improved cluster structure, as suggested by the silhouette score, make it a worthwhile approach to consider for clustering in this case.")
```
